```python
#samodzielne wpisywanie danych
#wykres radarowy
#wariant I
properties = ['Wytrzymałość', 'Twardość', 'Przewodność\ncieplna', 'Gęstość', 'Moduł Younga']
materials = ['Stal', 'Aluminium', 'Miedź']
values = np.array([
    [550, 150, 50, 7.8, 210],
    [300, 60, 220, 2.7, 70],
    [220, 50, 400, 8.96, 130]
])
```

```python
#okreslenie czy bedzie cale kolko czy czesc kolka
angles = np.linspace(0,2 * np.pi, len(properties) + 1)
fig = plt.figure(figsize=(5,5))
ax = plt.subplot(polar=True)

colors = ['blue', 'orange', 'green']

#zamykanie wielokąta
vals1 = np.concatenate((values[0], [values[0][0]]))
vals2 = np.concatenate((values[1], [values[0][1]]))
vals3 = np.concatenate((values[2], [values[0][2]]))

#rysunek
ax.plot(angles, vals1, color=colors[0], marker='o', label=materials[0])
ax.plot(angles, vals2, color=colors[1], marker='o', label=materials[1])
ax.plot(angles, vals3, color=colors[2], marker='o', label=materials[2])

plt.fill(angles, vals1, alpha=0.1)
plt.fill(angles, vals2, alpha=0.1)
plt.fill(angles, vals3, alpha=0.1)

ax.set_xticks(angles[:-1]) #osie
ax.set_xticklabels(properties)
ax.set_title('Właściwości materiałów', y = 1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
#wariant II
categories = ['hardness', 'toughness', 'density', 'conductivity', 'corrosion resistance']
materials2 = ['steel', 'aluminium', 'copper']
colors = ['red', 'blue', 'green']
values2 = np.array([
    [8, 7, 9, 5, 6],
    [5, 9, 4, 8, 7],
    [6, 6, 8, 7, 9]
])
```

```python
fig, axes = plt.subplots(1, 3, subplot_kw = {'projection': 'polar'}, figsize=(15, 5))

for i, ax in enumerate(axes):
    vals = np.concatenate((values2[i], [values2[i][0]]))
    
    ax.plot(angles, vals, color=colors[i], marker='o', linewidth=2)
    ax.set_title(materials2[i], size=14, pad = 20)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, size=10)
plt.suptitle('Material properties comparison', size=16, y=1.05)
plt.tight_layout()
plt.show()
```
