```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X, y_true = make_blobs(n_samples=300, centers = 5, cluster_std = 3.0, random_state=42)
plt.scatter(X[:,0], X[:,1], s=20, c='grey')
plt.show()
```

```python
y_true
```

```python
kmeans = KMeans(n_clusters = 7, n_init = 10, init='random', random_state=42)
kmeans.fit(X) #uczenie modelu! -> szukaj skupień 
y_kmeans = kmeans.labels_
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='black', s=80, alpha=0.5)
plt.show()
```

```python
print('SSE:', kmeans.inertia_)
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
X, y = make_moons(n_samples=200, noise=0.05, random_state=42)
plt.scatter(X[:,0], X[:,1], s=20, c='grey')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, n_init=10, init='random', random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.labels_
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=30, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='black', s=80, alpha=0.5)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
y_dbscan = dbscan.labels_
plt.scatter(X[:,0], X[:,1], c=y_dbscan, s=30, cmap='viridis')
plt.show()
```

```python
file = 'ph-data.csv'
df = pd.read_csv(file)
df.head()
```

```python
X = df.values[:,0:3]
y_true = df.values[:,3]
kmeans = KMeans(n_clusters=15, n_init=10, init='random', random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.labels_
plt.scatter(X[:,2], X[:,1], c=y_kmeans, s=30, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,2], centers[:,1], c='black', s=80, alpha=0.5)
plt.show()
```
